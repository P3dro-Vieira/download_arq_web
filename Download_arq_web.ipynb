```python
# 1. IMPORTAR O QUE PRECISO

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


# 2. ABRINDO O NAVEGADOR

driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.investing.com/currencies/usd-brl-historical-data")
time.sleep(2)


# 3. CLICAR EM ACEITAR OS COOKIES (IDENTIFICAR PELA ID), E ESPERAR A PAGINA CARREGAR,
# POR ISSO O WebDriverWait.

elemento = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.ID, "onetrust-accept-btn-handler")))
time.sleep(1)
elemento.click() #Clica no botão aceitar


# 4. CLICAR PARA NÃO TRADUZIR A PÁGINA

fechar_trad = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="topAlertBarContainer"]/div/div[3]/a[1]')))
time.sleep(1)
fechar_trad.click()


# 5. FECHAR A ABA DO CHAT

fechar_chat =  WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="leftColumn"]/div[13]/div[1]/div')))
time.sleep(1)
fechar_chat.click()


# 6. CLICANDO EM DOWNLOAD, ABRE UMA JANELA DE LOGIN

down_1 = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="column-content"]/div[4]/div/a')))

time.sleep(1)
down_1.click()


# 7. DIGITANDO EMAIL, SENHA E LOGANDO

email = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.ID, 'loginFormUser_email')))
time.sleep(1)
email.send_keys("seu_email")


pwd = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="loginForm_password"]')))
time.sleep(1)
pwd.send_keys("sua_senha")


sign = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="signup"]/a')))
time.sleep(1)
sign.click()


# 8. FAZENDO O DOWNLOAD (REPETI O CÓDIGO NA ETAPA 6)

down_1 = WebDriverWait(driver, 5).until(
EC.presence_of_element_located((By.XPATH, '//*[@id="column-content"]/div[4]/div/a')))
time.sleep(1)
down_1.click()


# 9. FECHANDO O NAVEGADOR
time.sleep(5)
driver.close()
```
